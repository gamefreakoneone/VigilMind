```python
import json

# Input and output file paths
input_file = r"oisd_nsfw_domainswild2.txt"   # your text file
output_file = r"blocklists.json" # JSON output

# Initialize JSON structure
data = {
    "blacklists": [],
    "graylists": [],
    "whitelists": []
}

# Read the text file and populate blacklist
with open(input_file, "r", encoding="utf-8") as f:
    for line in f:
        url = line.strip()
        if url:  # skip empty lines
            data["blacklists"].append({
                "url": url,
                "reason": "pornography"
            })

# Write to JSON file with pretty formatting
with open(output_file, "w", encoding="utf-8") as f:
    json.dump(data, f, indent=4, ensure_ascii=False)

print(f"✅ Blocklist converted and saved to {output_file}")
```
